```python
# importando bibliotecas
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import os
```

```python
#importando e exibindo a imagem original
imagem_path = '/content/drive/MyDrive/MESTRADO/Digital Image Processing/commander_eileen_collins.jpg'

imagem = plt.imread(imagem_path)
plt.imshow(imagem)
plt.axis('off')
plt.show()
```

## **Quantização da Imagem**

```python
# convertendo a imagem em float32 para as operações matemáticas seguintes
imagem = imagem.astype('float32')
imagem_cinza = []

# obtendo a altura e largura da imagem original
altura, largura = imagem.shape[:2]

# somando o produto dos canais RGB por 0.30, 0.59 e 0.11 respectivamente
for i in range(0, altura):
    for j in range(0, largura):
        soma = imagem[i][j][0]*0.30 + imagem[i][j][1]*0.59 + imagem[i][j][2]*0.11
        imagem_cinza.append(soma)
        soma = 0

# transformando a imagem em array e reordenando as suas dimensoes
imagem_cinza = np.array(imagem_cinza)
imagem_cinza = imagem_cinza.reshape(altura, largura)

# exibindo a imagem em escala de cinza
plt.imshow(imagem_cinza, cmap='gray')
plt.axis('on')
plt.show()
```

## **Amostragem em 4 níveis de cinza**

```python
quantizacao_niveis = 4

imagem_quantizada = np.floor_divide(imagem_cinza, 256 / quantizacao_niveis) * (256 / quantizacao_niveis)

plt.imshow(imagem_quantizada, cmap='gray')
plt.axis('off')
plt.title('4 níveis')
```

```python
# Defina os 4 níveis de cinza (valores de intensidade)
niveis_de_cinza = np.unique(imagem_quantizada)

# Crie um array de posições no eixo x
posicoes_x = np.arange(len(niveis_de_cinza))

# Crie um gráfico de escadas
plt.figure(figsize=(4, 4))
plt.step(posicoes_x, niveis_de_cinza, where='mid', color='b')
plt.xticks(posicoes_x, [str(elemento) for elemento in list(range(0, len(niveis_de_cinza)))])
plt.xlabel('Níveis')
plt.ylabel('Valores de Cinza')
plt.title('4 níveis')

plt.tight_layout()
plt.show()
```
