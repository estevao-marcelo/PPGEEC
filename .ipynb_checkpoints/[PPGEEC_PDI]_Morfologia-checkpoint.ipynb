```python
# importanto bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, morphology
```

```python
# carregando a imagem
img = io.imread('imagem_morfologia.png', as_gray=True) 
```

```python
# Binarizando a imagem
binarizada_img = img > 0.5
```

```python
# Aplicando erosão para remover ruído pequeno
selem = morphology.disk(3) # Elemento estruturante (disco) para a operação de erosão
erodida_img = morphology.erosion(binarizada_img, selem)
```

```python
# Aplicando dilatação para restaurar a forma dos objetos
dilatada_img = morphology.dilation(erodida_img, selem)
```
